```python
import numpy as np
from tqdm import tqdm

from WindyGridworld import windygridworld
import sarsa_agent as agent
```

# Windy GridWorld

## Initialize Environment

```python
env = windygridworld(king=False, stop=False, stochastic=False)
```

```python
env.render()
```

## Define value table $Q(s, a)$

### Create the $Q(s, a)$ table

```python
action_values = agent.initialize_action_values((env.observation_space[0],env.observation_space[1],env.action_space))
```

```python
np.shape(action_values)
```

## Test the policy $\pi(s)$

```python
for i in range (10):
    action = agent.policy(env.state, action_values, env.action_space, epsilon=0.1)
    print(f"Action taken in state {env.state} : {action}")
```

## Run On-Policy TD Control algorithm (SARSA)

```python
agent.on_policy_td_control(env, action_values, 170, epsilon=0.2)
```

## Listing Policy after learning

```python
for row in range(np.shape(action_values)[0]):
    for col in range (np.shape(action_values)[1]):
        action = agent.policy(np.array([row, col]), action_values, env.action_space)
        print(f"Action for state ({row}, {col}): {env.action_labels[action]}")
```

## Test Agent

```python
for a in range(5):
    agent.run_episode(env, action_values, epsilon=0.)
```

```python
agent.plot_results(env.observation_space, env.actions, action_values)
```
